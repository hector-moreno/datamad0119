```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np
from bs4 import BeautifulSoup

# proporcionada por la documentaci√≥n de la API: https://rickandmortyapi.com/documentation
url = 'https://www.quandl.com/api/v3/datasets/WIKI/AAPL.json'
```

```python
response = requests.get(url)
results = response.json()
results
```

```python
["Date","Open","High","Low","Close","Volume","Ex-Dividend","Split Ratio","Adj. Open","Adj. High","Adj. Low","Adj. Close","Adj. Volume"]
data = pd.DataFrame(results)
datel = []
openl = []
highl = []
lowl = []
closel = []
resultl = []

for list in data['dataset'][3][:1185]:
    datel.append(list[0])
    openl.append(list[1])
    highl.append(list[2])
    lowl.append(list[3])
    closel.append(list[4])


df = pd.DataFrame({'Date': datel, 'Open' : openl, 'High' : highl, 'Low' : lowl, 'Close' : closel})

df
```

```python
url2 = 'https://www.ft.com/stream/a39a4558-f562-4dca-8774-000246e6eebe?format=&page=' 

dates = []
headlines = []
for i in range(16,143): 
    path = url2 + str(i)
    html = requests.get(path).content
    soup = BeautifulSoup(html, "lxml")
    text = soup.find_all('li',{'class':'o-teaser-collection__item o-grid-row'})
    for i in soup.findAll('time'):
        if i.has_attr('datetime'):
            dates.append(i['datetime'][0:10])
    for i in soup.find_all('a',{'class':'js-teaser-heading-link'}):
        headlines.append(i.text)

dfHead = pd.DataFrame({'Fecha': dates, 'Titulares' : headlines})
```

```python
dfHead = pd.DataFrame({'Date': dates, 'Titulares' : headlines})
dfHead.head(10)
```

```python
dfHead = dfHead.groupby('Date', sort=False, as_index=False).agg({'Titulares': '| '.join})
dfHead.head(10)
```
